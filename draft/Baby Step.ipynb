```python
import theano.tensor as T
import numpy as np
from theano import function
```

```python
x = T.dscalar('x')
y = T.dscalar('y')

z = x + y
f = function([x, y], z)
```

```python
f(1,2)
```

```python
grad_x = T.grad(z,x)
```

```python
f_backward = function([x, y], grad_x)
```

```python
f_backward(1,2)
```

```python
x , w1 = T.dmatrices('x', 'w1')
b1 = T.vector('b1')

output_1 = x.dot(w1) + b1

f = function([x, w1, b1], output_1)



input_lenght = 100
input_dim = 3
output_dim = 2
output_dim_2 = 4

w2 = T.dmatrix('w2')
b2 = T.vector('b2')

output_2 = output_1.dot(w2) + b2
f2 = function([x, w1, b1, w2, b2], output_2)


x = np.random.rand(input_lenght, input_dim)
w = np.random.rand(input_dim, output_dim)
b = np.random.rand(output_dim)
w2 = np.random.rand(output_dim, output_dim_2)
b2 = np.random.rand(output_dim_2)

print(b)
# f(x,w, b)
f2(x, w, b, w2, b2)



```
